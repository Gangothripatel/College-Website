```python
import zipfile
import os
import time

# Define paths to the zip files
val_zip_path = '/content/drive/MyDrive/dataset/coco_dataset/images/val/val2017.zip'
train_zip_path = '/content/drive/MyDrive/dataset/coco_dataset/images/train/train2017 (1).zip'
annotations_zip_path = '/content/drive/MyDrive/dataset/coco_dataset/annotations/annotations_trainval2017 (1).zip'

# Define the extraction paths
val_extract_path = '/content/drive/MyDrive/dataset/coco_dataset/images/val/'
train_extract_path = '/content/drive/MyDrive/dataset/coco_dataset/images/train/'
annotations_extract_path = '/content/drive/MyDrive/dataset/coco_dataset/annotations/'

# Function to unzip a file
def unzip_file(zip_path, extract_path):
    with zipfile.ZipFile(zip_path, 'r') as zip_ref:
        zip_ref.extractall(extract_path)
    print(f"Extracted {zip_path} to {extract_path}")

# Unzipping the validation images
unzip_file(val_zip_path, val_extract_path)
time.sleep(5)  # Wait for 5 seconds before the next extraction

# Unzipping the training images
unzip_file(train_zip_path, train_extract_path)
time.sleep(5)  # Wait for 5 seconds before the next extraction

# Unzipping the annotations
unzip_file(annotations_zip_path, annotations_extract_path)
```
